```python
from scipy import cluster
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering


tfid = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfid.fit_transform(df['name'])

from sklearn.decomposition import TruncatedSVD

lsa = TruncatedSVD(n_components=100)
X_lsa = lsa.fit_transform(tfidf_matrix)

explained_variance = lsa.explained_variance_ratio_.sum()

X_lsa.shape, explained_variance
agg_clustering_lsa = AgglomerativeClustering(n_clusters=10)
agg_clusters_lsa = agg_clustering_lsa.fit_predict(X_lsa)

num_clusters_lsa = len(set(agg_clusters_lsa))

df['name'] = agg_clusters_lsa
print(num_clusters_lsa)

df.head()
```

#### Dealing with Category

```python
label_encoder = LabelEncoder()
df['category'] = label_encoder.fit_transform(df['category'])
df.head()
```

### Target Encoder for item_condition

```python
encoder = TargetEncoder()

encoder.fit(df['item_condition'], df['price_log'])
df['item_condition'] = encoder.transform(df['item_condition'])

encoder.fit(df['brand'], df['price_log'])
df['brand'] = encoder.transform(df['brand'])

encoder.fit(df['located_in'], df['price_log'])
df['located_in'] = encoder.transform(df['located_in'])

# df.head()
```
